```python
import os
import re
from collections import defaultdict

# Define regex patterns to extract the X part from the two filename formats
pattern1 = r'frame_(\w+)_\w+_jpg\.rf\.\w+\.jpg'  # For frame_X_Y_jpg.rf.Z.jpg
pattern2 = r'video_(\w+)_mp4-\w+_jpg\.rf\.\w+\.jpg'  # For video_X_mp4-Y_jpg.rf.Z.jpg

# Directory containing your files
directory = 'datasets/exp/full/images/'
directory_labels = 'datasets/exp/full/labels/'

# Dictionary to hold files grouped by X value
file_groups = defaultdict(list)

# Iterate through all files in the directory
for filename in os.listdir(directory):
    # Try to match both patterns
    match1 = re.match(pattern1, filename)
    match2 = re.match(pattern2, filename)
    
    # If it matches the first pattern
    if match1:
        X_value = match1.group(1)
        file_groups[X_value].append(filename)
    # If it matches the second pattern
    elif match2:
        X_value = match2.group(1)
        file_groups[X_value].append(filename)



# Print the grouped files
gg = 0
Xs = sorted(list(file_groups.keys()))

for x in Xs:
    print(x)

print(gg)
```

#### Delete previous dataset

```python
def nuke():
    root_dir = "datasets/exp/"
    main_dataset_dir = "datasets/exp/full/"
    for root, dirs, files in os.walk(root_dir, topdown=False):

        if root == root_dir:
            print(f"Skipping files in the root directory: {root}")
            continue

        # Skip deletion if it's the excluded child directory
        if main_dataset_dir in root:
            print(f"Skipping files in: {root}")
            continue
        
        # Delete files in the current directory
        for file_name in files:
            file_path = os.path.join(root, file_name)
            print(f"Deleting file: {file_path}")
            os.remove(file_path)

# nuke()
```

```python
Xs = sorted(list(file_groups.keys()))
```

#### Prepare test set

```python
import shutil 

destination_directory = "datasets/exp/test/images/"
destination_directory_label = "datasets/exp/test/labels/"

test_x = []
count = 100
for x in Xs:
    if len(x) < 4 and count > 0:  # len<4 means this is a sample from yt collected data
        count -= len(file_groups[x])
        test_x.append(x)

for x in test_x:
    for file_to_copy in file_groups[x]:
        source_file_path = os.path.join(directory, file_to_copy)
        source_label_path = os.path.join(directory_labels, file_to_copy.replace(".jpg", ".txt"))
            
        # Full destination path where the file will be copied
        destination_file_path = os.path.join(destination_directory, file_to_copy)
        destination_label_path = os.path.join(destination_directory_label, file_to_copy.replace(".jpg", ".txt"))
        
        # Copy the file to the destination folder
        shutil.copy(source_file_path, destination_file_path)
        shutil.copy(source_label_path, destination_label_path)
```

#### Prepare val set

```python
import shutil 

destination_directory = "datasets/exp/valid/images/"
destination_directory_label = "datasets/exp/valid/labels/"

val_x = []
count = 150
for x in Xs:
    if "20240923" in x and count > 0 :  # 20240923 means its collected from zindabajar parking lots
        count -= len(file_groups[x])
        val_x.append(x)

for x in val_x:
    for file_to_copy in file_groups[x]:
        source_file_path = os.path.join(directory, file_to_copy)
        source_label_path = os.path.join(directory_labels, file_to_copy.replace(".jpg", ".txt"))
            
        # Full destination path where the file will be copied
        destination_file_path = os.path.join(destination_directory, file_to_copy)
        destination_label_path = os.path.join(destination_directory_label, file_to_copy.replace(".jpg", ".txt"))
        
        # Copy the file to the destination folder
        shutil.copy(source_file_path, destination_file_path)
        shutil.copy(source_label_path, destination_label_path)
```

#### Prepare train set

```python
import shutil 

destination_directory = "datasets/exp/train/images/"
destination_directory_label = "datasets/exp/train/labels/"

for x in Xs:
    
    if x in val_x or x in test_x:
        continue

    for file_to_copy in file_groups[x]:
        source_file_path = os.path.join(directory, file_to_copy)
        source_label_path = os.path.join(directory_labels, file_to_copy.replace(".jpg", ".txt"))
            
        # Full destination path where the file will be copied
        destination_file_path = os.path.join(destination_directory, file_to_copy)
        destination_label_path = os.path.join(destination_directory_label, file_to_copy.replace(".jpg", ".txt"))
        
        # Copy the file to the destination folder
        shutil.copy(source_file_path, destination_file_path)
        shutil.copy(source_label_path, destination_label_path)
```
